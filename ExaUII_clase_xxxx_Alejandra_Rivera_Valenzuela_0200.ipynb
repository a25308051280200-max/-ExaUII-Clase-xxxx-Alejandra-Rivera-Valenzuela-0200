### Crear la clase "Puente", con 3 atributos (nombre, longitud, tipoConstruccion), 2 funciones leerDatos(), mostrarDatos() y 2 objetos (puente1, puente2)

### Que sea interactivo --> los input

### No usar while true, if, for, listas y diccionario

![texto del vínculo](https://raw.githubusercontent.com/a25308051280200-max/-ExaUII-Clase-xxxx-Alejandra-Rivera-Valenzuela-0200/refs/heads/main/infoexaIIclassxxx.png)

```python
class Puente:
    def __init__(self, nombre, longitud, tipoConstruccion):
        self.nombre = nombre
        self.longitud = longitud
        self.tipoConstruccion = tipoConstruccion

    def leerDatos(self):
      self.nombre = input("Ingresa el nombre Nombre: ")
      self.longitud = input("Ingresa la Longitud: ")
      self.tipoConstruccion = input("Ingresa el tipoConstruccion: ")

    def mostrarDatos(self):
        print("Nombre:", self.nombre)
        print("Longitud:", self.longitud)
        print("Tipo de Construcción:", self.tipoConstruccion)

# creando los objetos puente1 y puente2
puente1 = Puente("", "", "")
puente2 = Puente("", "", "")

print("Puente Numero 1")
puente1.leerDatos()
puente1.mostrarDatos()
print("<3-<3-<3-<3-<3-<3-<3")

print("Puente Numero 2")
puente2.leerDatos()
puente2.mostrarDatos()
print("<3-<3-<3-<3-<3-<3-<3")


print("Software desarollado por Alejandra Rivera Valenzuela NL=45")
```
